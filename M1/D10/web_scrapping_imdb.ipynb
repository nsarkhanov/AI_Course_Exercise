```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=war&sort=user_rating,desc')
soup = BeautifulSoup(page.content, 'html.parser')
films_lists=soup.find('div',class_="lister list detail sub-list")

print(films_lists)
```
